### Step2. 2023년 2024년 관람객 추이 살펴보기

```python
# 1. 2023년과 2024년에서 필요한 컬럼만 추출
df_art_view_2023 = df_art_2023[['FCLTY_NM', 'VIEWNG_NMPR_CO']].copy()
df_art_view_2024 = df_art_2024[['FCLTY_NM', 'VIEWNG_NMPR_CO']].copy()

# 2. 컬럼명 변경
df_art_view_2023.rename(columns={
    'FCLTY_NM': '2023_시설명',
    'VIEWNG_NMPR_CO': '2023_관람인원수'
}, inplace=True)

df_art_view_2024.rename(columns={
    'FCLTY_NM': '2024_시설명',
    'VIEWNG_NMPR_CO': '2024_관람인원수'
}, inplace=True)
```

```python
# 1. 조인을 위한 임시 키 생성
df_art_view_2023['시설명_공통'] = df_art_view_2023['2023_시설명']
df_art_view_2024['시설명_공통'] = df_art_view_2024['2024_시설명']

# 2. inner join 수행 (2024년 데이터 기준으로)
df_joined = pd.merge(
    df_art_view_2024,
    df_art_view_2023,
    on='시설명_공통',
    how='inner',
    suffixes=('_2024', '_2023')
)

# 3. 원래 원하는 컬럼만 남기고 정리
df_art_viewing = df_joined[[
    '2024_시설명', '2024_관람인원수', '2023_관람인원수'
]]
```

```python
df_art_viewing.shape
```

```python
df_art_viewing.info()
```

```python
# '미제출'인 행 필터링
misubmitted_rows = df_art_viewing[df_art_viewing['2024_관람인원수'] == '미제출']

# 결과 출력
print(misubmitted_rows)
```

```python
df_art_viewing['2024_관람인원수'] = df_art_viewing['2024_관람인원수'].replace('미제출', np.nan)
```

```python
df_art_viewing.info()
```

```python
# float형으로 변환
df_art_viewing['2024_관람인원수'] = df_art_viewing['2024_관람인원수'].astype(float)
```

```python
# 조건: 두 열 중 하나라도 NaN인 경우 필터링
null_rows = df_art_viewing[
    df_art_viewing['2024_관람인원수'].isna() | df_art_viewing['2023_관람인원수'].isna()
]

# 결과 출력
print(null_rows)
```

```python
import numpy as np

# 전, 후 평균을 계산하는 함수 정의
def fillna_with_neighbor_avg(series):
    series_filled = series.copy()
    
    for i in range(1, len(series) - 1):
        if pd.isna(series[i]):
            prev_val = series[i - 1]
            next_val = series[i + 1]
            if not pd.isna(prev_val) and not pd.isna(next_val):
                series_filled[i] = (prev_val + next_val) / 2
    
    return series_filled

df_art_viewing['2024_관람인원수'] = fillna_with_neighbor_avg(df_art_viewing['2024_관람인원수'])
df_art_viewing['2023_관람인원수'] = fillna_with_neighbor_avg(df_art_viewing['2023_관람인원수'])
```

```python
# 남은 NaN을 전체 평균으로 대체
df_art_viewing['2024_관람인원수'] = df_art_viewing['2024_관람인원수'].fillna(df_art_viewing['2024_관람인원수'].mean())
df_art_viewing['2023_관람인원수'] = df_art_viewing['2023_관람인원수'].fillna(df_art_viewing['2023_관람인원수'].mean())
```

```python
df_art_viewing.info()
```

```python
# 관람객 증감 칼럼 추가
df_art_viewing['관람객_증감'] = (
    df_art_viewing['2024_관람인원수'].astype(int) - df_art_viewing['2023_관람인원수'].astype(int)
)
```

```python
# 관람객 증감률 계산 (단위: %), 소수점 1자리 반올림
df_art_viewing['관람객_증감률'] = round(
    (
        (df_art_viewing['2024_관람인원수'] - df_art_viewing['2023_관람인원수']) 
        / df_art_viewing['2023_관람인원수']
    ) * 100, 1
)
```

```python
from matplotlib.patches import Patch
```

```python
# 0인 값 제거 후 색상 지정
df_filtered = df_art_viewing[df_art_viewing['관람객_증감률'] != 0].copy()
df_filtered['color'] = df_filtered['관람객_증감률'].apply(lambda x: 'red' if x > 0 else 'blue')

# 오름차순 정렬
df_sorted = df_filtered.sort_values(by='관람객_증감률')

# 그래프 그리기
plt.figure(figsize=(25, 20))
plt.bar(
    x=range(len(df_sorted)),
    height=df_sorted['관람객_증감률'],
    color=df_sorted['color']
)

# 기준선 추가
plt.axhline(0, color='gray', linestyle='--', linewidth=1)

# x축 라벨 및 눈금 완전 제거
plt.xticks([], [])

# 범례 추가
custom_legend = [
    Patch(color='red', label='관람객 증가'),
    Patch(color='blue', label='관람객 감소')
]
plt.legend(handles=custom_legend, title='증감률 상태', loc='upper right')

# 기타 라벨 및 제목
plt.xlabel('미술관 시설 (숨김 처리됨)')
plt.ylabel('관람객 증감률 (%)')
plt.title('미술관별 관람객 증감률 (x축 라벨 숨김)')

plt.tight_layout()
plt.show()
```

```python
# 음수 개수 (관람객 감소)
num_negative = (df_art_viewing['관람객_증감률'] < 0).sum()

# 양수 개수 (관람객 증가)
num_positive = (df_art_viewing['관람객_증감률'] > 0).sum()

# 변화 없음 (증감률 == 0)
num_zero = (df_art_viewing['관람객_증감률'] == 0).sum()

print(f"관람객 감소(음수) 개수: {num_negative}")
print(f"관람객 증가(양수) 개수: {num_positive}")
print(f"변화 없음(0) 개수: {num_zero}")
```
