- 'BULD_AR_VALUE' 기준 그룹별로 평균을 구하여 'DATA_CO'의 null값 처리하기
- 'BULD_AR_VALUE' 가 Null인 Data는 그룹별 집계할 때 제외

```python
# 1. BULD_AR_VALUE가 null이 아닌 행에 대해서만 분위 그룹 생성
df_ArtData_2024.loc[df_ArtData_2024['BULD_AR_VALUE'].notnull(), 'BULD_AR_VALUE_group'] = \
    pd.qcut(df_ArtData_2024.loc[df_ArtData_2024['BULD_AR_VALUE'].notnull(), 'BULD_AR_VALUE'], q=4, labels=False)

# 2. 분위 그룹별 DATA_CO 평균 계산
group_means_DATA_CO = df_ArtData_2024.groupby('BULD_AR_VALUE_group')['DATA_CO'].transform('mean')

# 3. DATA_CO가 null이고, 그룹이 있는 경우에만 평균으로 채우기
df_ArtData_2024['DATA_CO'] = df_ArtData_2024.apply(
    lambda row: group_means_DATA_CO[row.name] if pd.isna(row['DATA_CO']) and pd.notna(row['BULD_AR_VALUE_group']) else row['DATA_CO'],
    axis=1
)

# 4. 임시 그룹 컬럼 삭제
df_ArtData_2024.drop(columns='BULD_AR_VALUE_group', inplace=True)
```

```python
df_ArtData_2024.info()
```

```python
# 1. DATA_CO에서 결측치가 남아 있는 행 인덱스 찾기
null_idx_DATA_CO = df_ArtData_2024[df_ArtData_2024['DATA_CO'].isnull()].index

# 2. 앞뒤 평균으로 대체 (하나만 남아 있다고 가정)
for idx in null_idx_DATA_CO:
    prev_val_DC = df_ArtData_2024.loc[idx - 1, 'DATA_CO'] if idx > 0 else np.nan
    next_val_DC = df_ArtData_2024.loc[idx + 1, 'DATA_CO'] if idx < len(df_ArtData_2024) - 1 else np.nan
    avg_val_DC = np.nanmean([prev_val_DC, next_val_DC])
    
    # 소수점 버림 후 대체
    df_ArtData_2024.at[idx, 'DATA_CO'] = np.floor(avg_val_DC) if not np.isnan(avg_val_DC) else np.nan
```

```python
# 이상치 제거용 함수 (IQR 기준)
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# 이상치 제거된 데이터 생성
df_clean_BULD = remove_outliers_iqr(df_ArtData_2024, 'BULD_AR_VALUE')

# 시각화
plt.figure(figsize=(14, 6))
sns.boxplot(x='CTPRVN_NM', y='BULD_AR_VALUE', data=df_clean_BULD, palette='Set3')

plt.title('CTPRVN_NM별 BULD_AR_VALUE 분포(Boxplot, 이상치 제외)', fontsize=15)
plt.xlabel('CTPRVN_NM', fontsize=8)
plt.ylabel('BULD_AR_VALUE (㎡)', fontsize=12)
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

- 시도명 기준으로 그룹을 묶어 평균을 계산한뒤 'BULD_AR_VALUE'의 null값을 처리했다.

```python
# 1. 이상치 제거 함수 (IQR 기준)
def remove_outliers_iqr_series(s):
    Q1 = s.quantile(0.25)
    Q3 = s.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return s[(s >= lower) & (s <= upper)]

# 2. 이상치 제거 후, CTPRVN_NM별 평균 계산 (딕셔너리 형태로 저장)
group_means_no_outliers = {}

for region, group in df_ArtData_2024.groupby('CTPRVN_NM'):
    clean_values = remove_outliers_iqr_series(group['BULD_AR_VALUE'].dropna())
    if not clean_values.empty:
        group_means_no_outliers[region] = clean_values.mean()
    else:
        group_means_no_outliers[region] = None  # 혹시 전부 이상치일 경우

# 3. null인 경우 해당 그룹의 이상치 제거 평균값으로 채우기
def fill_buld_with_group_mean(row):
    if pd.isna(row['BULD_AR_VALUE']):
        return group_means_no_outliers.get(row['CTPRVN_NM'], row['BULD_AR_VALUE'])
    else:
        return row['BULD_AR_VALUE']

df_ArtData_2024['BULD_AR_VALUE'] = df_ArtData_2024.apply(fill_buld_with_group_mean, axis=1)
```

```python
# 1. 공립/사립만 추출
mask_FLAG = df_ArtData_2024['FLAG_NM'].isin(['공립', '사립'])

# 2. 평균 계산 (공립/사립 각각)
group_means_mVP = df_ArtData_2024[mask_FLAG].groupby('FLAG_NM')['VIEWNG_PRICE'].mean()

# 3. 결측값 처리: FLAG_NM이 공립 또는 사립이고 VIEWNG_PRICE가 null인 경우 → 해당 그룹 평균으로 채우기
def fill_viewng_price(row):
    if pd.isna(row['VIEWNG_PRICE']) and row['FLAG_NM'] in ['공립', '사립']:
        return group_means_mVP[row['FLAG_NM']]
    return row['VIEWNG_PRICE']

df_ArtData_2024['VIEWNG_PRICE'] = df_ArtData_2024.apply(fill_viewng_price, axis=1)
```

```python
# 1. '미제출'을 np.nan으로 변환
df_ArtData_2024['VIEWNG_NMPR_CO'] = df_ArtData_2024['VIEWNG_NMPR_CO'].replace('미제출', np.nan)

# 2. 타입을 float으로 변환 (문자형이었을 경우)
df_ArtData_2024['VIEWNG_NMPR_CO'] = pd.to_numeric(df_ArtData_2024['VIEWNG_NMPR_CO'], errors='coerce')

# 3. 앞뒤 평균으로 채우기 (소수점은 버림)
def fill_with_avg(row_idx):
    prev_val = df_ArtData_2024.loc[row_idx - 1, 'VIEWNG_NMPR_CO'] if row_idx > 0 else np.nan
    next_val = df_ArtData_2024.loc[row_idx + 1, 'VIEWNG_NMPR_CO'] if row_idx < len(df_ArtData_2024) - 1 else np.nan
    avg_val = np.nanmean([prev_val, next_val])
    return np.floor(avg_val) if not np.isnan(avg_val) else np.nan

# 4. null 값 위치 확인 및 평균 대체 적용
for idx, val in df_ArtData_2024['VIEWNG_NMPR_CO'].items():
    if pd.isna(val):
        df_ArtData_2024.at[idx, 'VIEWNG_NMPR_CO'] = fill_with_avg(idx)

# 5. 타입 정수형으로 변경 (선택)
df_ArtData_2024['VIEWNG_NMPR_CO'] = df_ArtData_2024['VIEWNG_NMPR_CO'].astype('Int64')  # 결측 가능 정수형
```
