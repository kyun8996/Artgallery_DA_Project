## Step2. 2024 미술관 데이터와 같이 null값 처리하기

- '2023_DATA_CO'칼럼 Null값 처리하기

```python
# 1. 숫자형으로 안전하게 변환 (변환 불가능한 값은 NaN)
df_art_mart['2023_DATA_CO'] = pd.to_numeric(df_art_mart['2023_DATA_CO'], errors='coerce')
```

```python
# 1. BULD_AR_VALUE가 null이 아닌 행에 대해서만 분위 그룹 생성
df_art_mart.loc[df_art_mart['2023_BULD_AR_VALUE'].notnull(), '2023_BULD_AR_VALUE_group'] = \
    pd.qcut(df_art_mart.loc[df_art_mart['2023_BULD_AR_VALUE'].notnull(), '2023_BULD_AR_VALUE'], q=4, labels=False)

# 2. 분위 그룹별 DATA_CO 평균 계산
group_means_DATA_CO = df_art_mart.groupby('2023_BULD_AR_VALUE_group')['2023_DATA_CO'].transform('mean')

# 3. DATA_CO가 null이고, 그룹이 있는 경우에만 평균으로 채우기
df_art_mart['2023_DATA_CO'] = df_art_mart.apply(
    lambda row: group_means_DATA_CO[row.name] if pd.isna(row['2023_DATA_CO']) and pd.notna(row['2023_BULD_AR_VALUE_group']) else row['2023_DATA_CO'],
    axis=1
)

# 4. 임시 그룹 컬럼 삭제
df_art_mart.drop(columns='2023_BULD_AR_VALUE_group', inplace=True)
```

```python
# 1. DATA_CO에서 결측치가 남아 있는 행 인덱스 찾기
null_idx_DATA_CO = df_art_mart[df_art_mart['2023_DATA_CO'].isnull()].index

# 2. 앞뒤 평균으로 대체 (하나만 남아 있다고 가정)
for idx in null_idx_DATA_CO:
    prev_val_DC = df_art_mart.loc[idx - 1, '2023_DATA_CO'] if idx > 0 else np.nan
    next_val_DC = df_art_mart.loc[idx + 1, '2023_DATA_CO'] if idx < len(df_art_mart) - 1 else np.nan
    avg_val_DC = np.nanmean([prev_val_DC, next_val_DC])
    
    # 소수점 버림 후 대체
    df_art_mart.at[idx, '2023_DATA_CO'] = np.floor(avg_val_DC) if not np.isnan(avg_val_DC) else np.nan
```

- '2023_BULD_AR_VALUE'칼럼 Null값 처리하기

```python
# 1. 이상치 제거 함수 (IQR 기준)
def remove_outliers_iqr_series(s):
    Q1 = s.quantile(0.25)
    Q3 = s.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return s[(s >= lower) & (s <= upper)]

# 2. 이상치 제거 후, CTPRVN_NM별 평균 계산 (딕셔너리 형태로 저장)
group_means_no_outliers = {}

for region, group in df_art_mart.groupby('CTPRVN_NM'):
    clean_values = remove_outliers_iqr_series(group['2023_BULD_AR_VALUE'].dropna())
    if not clean_values.empty:
        group_means_no_outliers[region] = clean_values.mean()
    else:
        group_means_no_outliers[region] = None  # 혹시 전부 이상치일 경우

# 3. null인 경우 해당 그룹의 이상치 제거 평균값으로 채우기
def fill_buld_with_group_mean(row):
    if pd.isna(row['2023_BULD_AR_VALUE']):
        return group_means_no_outliers.get(row['CTPRVN_NM'], row['2023_BULD_AR_VALUE'])
    else:
        return row['2023_BULD_AR_VALUE']

df_art_mart['2023_BULD_AR_VALUE'] = df_art_mart.apply(fill_buld_with_group_mean, axis=1)
```

```python
# 국립현대미술관 홈페이지 참조
# 사전 정의된 미술관별 관람료 (단위: 원)
price_map = {
    '국립현대미술관(과천관)': 3000,
    '국립현대미술관(서울관)': 7000,
    '국립현대미술관(덕수궁관)': 0,
    '국립현대미술관 미술품수장센터(청주관)': 2000
}

# 각 항목에 대해 조건부로 null인 VIEWNG_PRICE 채우기
for name, price in price_map.items():
    df_art_mart.loc[
        (df_art_mart['2024_FCLTY_NM'] == name) &
        (df_art_mart['2023_VIEWNG_PRICE'].isna()),
        '2023_VIEWNG_PRICE'
    ] = price
```

```python
# 1. 공립/사립만 추출
mask_FLAG = df_art_mart['FLAG_NM'].isin(['공립', '사립'])

# 2. 평균 계산 (공립/사립 각각)
group_means_mVP = df_art_mart[mask_FLAG].groupby('FLAG_NM')['2023_VIEWNG_PRICE'].mean()

# 3. 결측값 처리: FLAG_NM이 공립 또는 사립이고 VIEWNG_PRICE가 null인 경우 → 해당 그룹 평균으로 채우기
def fill_viewng_price(row):
    if pd.isna(row['2023_VIEWNG_PRICE']) and row['FLAG_NM'] in ['공립', '사립']:
        return group_means_mVP[row['FLAG_NM']]
    return row['2023_VIEWNG_PRICE']

df_art_mart['2023_VIEWNG_PRICE'] = df_art_mart.apply(fill_viewng_price, axis=1)
```

- '2023_ARTGR_EMP_CO'칼럼의 Null값 처리하기

```python
# 1. 이상치 제거 함수 (IQR 기준)
def remove_outliers_iqr(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return series[(series >= lower) & (series <= upper)]

# 2. null이 아닌 값 중 이상치를 제거한 평균 계산
artgr_series_clean = remove_outliers_iqr(df_art_mart['2023_ARTGR_EMP_CO'].dropna())
mean_artgr_emp = artgr_series_clean.mean()

# 3. null 값을 평균으로 대체
df_art_mart['2023_ARTGR_EMP_CO'] = df_art_mart['2023_ARTGR_EMP_CO'].fillna(mean_artgr_emp)
```
